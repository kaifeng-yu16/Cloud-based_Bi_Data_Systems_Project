```
import pandas as pd

df = pd.read_csv("/dbfs/FileStore/shared_uploads/ky99@duke.edu/caffeine.csv")
```

### Preprocess data

```
df.head()
```

```
df.rename(
    columns = {
        "Volume (ml)": "volume",
        "Calories": "calories",
        "Caffeine (mg)": "caffeine",
    },
    inplace=True,
)
```

```
df.drop("drink", axis=1, inplace=True)
df.head()
```

```
df.to_csv("/dbfs/FileStore/shared_uploads/ky99@duke.edu/caffeine_processed.csv", index=False)
```

### Visualization

```
import matplotlib.pyplot as plt
import seaborn as sns
```

```
g=sns.catplot(x='type', kind='count', data=df)
g.fig.set_figheight(5)
g.fig.set_figwidth(10)
plt.show()
```

```
df['caffeine/ml']=df['caffeine']/df['volume']
g=sns.pairplot(df, hue='type')
g.fig.set_figheight(10)
g.fig.set_figwidth(20)
plt.show()
```

```
df['calories/100ml'] = df['calories']*100/df['volume']
df['caffeine/100ml'] = df['caffeine']*100/df['volume']

plt.figure(figsize=(15, 6))
sns.barplot(x='type',y='calories/100ml',data=df)
plt.title('Beverage Type vs Calories')
plt.xlabel('Beverage Type')
plt.ylabel('Calories')
plt.show()

plt.figure(figsize=(15, 6))
sns.barplot(x='type',y='caffeine/100ml',data=df)
plt.title('Beverage Type vs Caffeine')
plt.xlabel('Beverage Type')
plt.ylabel('Caffeine')
plt.show()
```
